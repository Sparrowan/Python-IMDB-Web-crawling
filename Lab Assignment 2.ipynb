## Q4. Create a new url list that contains the proper urls for top 10 movies. Call it `top10_urls_fixed`.

YOUR ANSWER for Q4 (Python code below)

```python
top10_urls_fixed = []

for t in top10_urls:
    t="".join(("http://www.imdb.com",t))
    top10_urls_fixed.append(t)
```

## Q5. Crawl the top 10 movies pages and store the pages (as text) as separate documents in your MongoDB

The document must have two keys: `url`, and `page`. For example: `{'url': 'https://www.imdb.com/title/tt0111161/', 'page': '<!DOCTYPE html><html xm...'}`

NOTE: IF YOU STORE THE PAGES AS BYTES (e.g., USING `movie_response.content`), LATER YOU NEED TO USE BYTES VERSION OF REGEX. IF YOU STORE THE PAGES AS TEXT (e.g., USING `movie_response.text`), LATER YOU NEED TO USE STRING VERSION OF REGEX. You can use either `.content` or `.text`.

YOUR ANSWER for Q5 (Python code below, use any number of cells )

```python
from pymongo import MongoClient

client = MongoClient("localhost", 27017)
db = client.get_database('lab2db')
```

```python
for t in top10_urls_fixed:
    response=requests.get(t)
    page_content=response.content
    movie = {"url": t,
            "page": page_content}
    movies = db.movies
    result = movies.insert_one(movie)
```
